# Year Features

####
The code generates two sets of subplots to visualize the relationship between various yearly features and two key metrics: Performance Rating and Attrition. For each feature containing "Yr" or "Year" in its name, the code plots the mean Performance Rating and mean Attrition rate across different values of that feature.
```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
fig, axs = plt.subplots(len(year_feature), 1, figsize=(10, 8))

for i, feature in enumerate(year_feature):
```
```python
    axs[i].plot(df.groupby(feature)['Attrition'].mean())
    axs[i].set_title(f"{feature} vs Attrition")
    axs[i].set_xlabel(feature)
    axs[i].set_ylabel('Attrition')

plt.tight_layout()
plt.show()
```

# Train-Test Split(For Attrition)

```python
from sklearn.model_selection import train_test_split

data=df.copy()

X_a = data.drop('Attrition', axis=1)
y_a = data['Attrition']

X_train_A, X_test_A, y_train_A, y_test_A = train_test_split(X_a, y_a, test_size=0.2, random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_a=StandardScaler()

X_train_a=scaler_a.fit_transform(X_train_A)
X_test_a=scaler_a.transform(X_test_A)
```

```python
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.model_selection import validation_curve
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
import random
import graphviz
import time
```

```python
def report(Y,prediction):
  return (classification_report(Y,prediction),confusion_matrix(Y,prediction),accuracy_score(Y,prediction))
```

# Logistic Regression

####
This code segment fits a Logistic Regression model to the training data (X_train_a, y_train_A), makes predictions on the test data (X_test_a), and computes evaluation metrics including the classification report, confusion matrix, and accuracy score. The logistic_model_report likely contains detailed metrics such as precision, recall, F1-score, and support for each class, while the logistic_model_confusion_matrix provides a summary of the model's performance across different classes, and logistic_model_accuracy stores the overall accuracy of the model.

```python
logistic_model=LogisticRegression(random_state=10)
```

```python
logistic_model.fit(X_train_a,y_train_A)
```

```python
predictions=logistic_model.predict(X_test_a)
```

```python
logistic_model_report,logistic_model_confusion_matrix,logistic_model_accuracy=report(y_test_A,predictions)
```

```python
print(logistic_model_report)
```

```python
print(logistic_model_confusion_matrix)
```

```python
logistic_model_accuracy
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=logistic_model_confusion_matrix,
                               display_labels=logistic_model.classes_)
disp.plot()
```
